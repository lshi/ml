```python
# import libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# load dataset
dataset = pd.read_csv('https://raw.githubusercontent.com/lshi/ml/main/data/position_salary.csv')
X = dataset.iloc[:, 1:2].values
y = dataset.iloc[:, 2].values
```

```python
# fit Linear Regression to dataset
from sklearn.linear_model import LinearRegression

lin_regr = LinearRegression()
lin_regr.fit(X, y)

# predict all data points
lin_regr_pred = lin_regr.predict(X)
```

```python
# visualise Linear Regression results
plt.scatter(X, y, color='blue')

plt.plot(X, lin_regr_pred, color='red')
plt.title('Linear Regression')
plt.xlabel('Position Level')
plt.ylabel('Salary')

plt.show()
```

```python
# fit Polynomial Regression to dataset
from sklearn.preprocessing import PolynomialFeatures

trans = PolynomialFeatures(degree=2) # try different degrees e.g. 2, 3, 4, 5,...
X_poly = trans.fit_transform(X)

trans.fit(X_poly, y)
lin_regr_poly = LinearRegression()
lin_regr_poly.fit(X_poly, y)

# predict all data points
lin_regr_pred_poly = lin_regr_poly.predict(X_poly)
```

```python
# visualise Polynomial Regression results
plt.scatter(X, y, color='blue')

plt.plot(X, lin_regr_pred_poly, color='red')
plt.title('Polynomial Regression')
plt.xlabel('Position Level')
plt.ylabel('Salary')

plt.show()
```

```python
# visualise Polynomial Regression results (for higher resolution & smoother curve)
X_grid = np.arange(min(X), max(X), 0.1)
X_grid = X_grid.reshape((len(X_grid), 1))

plt.scatter(X, y, color = 'blue')

plt.plot(X_grid, lin_regr_poly.predict(trans.fit_transform(X_grid)), color = 'red')
plt.title('Polynomial Regression')
plt.xlabel('Position Level')
plt.ylabel('Salary')

plt.show()
```

```python
# predict a new result with Linear Regression
lin_regr.predict(np.array([[7]]))
```

```python
# predict a new result with Polynomial Regression
lin_regr_poly.predict(trans.fit_transform(np.array([[7]])))
```
